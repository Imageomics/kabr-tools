```python
import os
import pandas as pd
import shutil
from natsort import natsorted
from tqdm import tqdm
```

```python
dataset_path = "/mnt/NVME/KABR"
new_dataset_path = "/mnt/NVME/KABR_SEGMENTED"

if not os.path.exists(new_dataset_path):
    os.makedirs(new_dataset_path)

if not os.path.exists(f"{new_dataset_path}/annotation"):
    os.makedirs(f"{new_dataset_path}/annotation")

if not os.path.exists(f"{new_dataset_path}/dataset/image"):
    os.makedirs(f"{new_dataset_path}/dataset/image")

train_df = pd.read_csv(f"{dataset_path}/annotation/train.csv", sep=" ")
val_df = pd.read_csv(f"{dataset_path}/annotation/val.csv", sep=" ")
segment_size = 90
```

```python
new_df = []
video_id = 1

for folder in tqdm(train_df["original_vido_id"].unique()):
    subset = train_df[train_df["original_vido_id"] == folder].reset_index(drop=True)
    
    number_of_subfolders = len(subset) // segment_size
    
    for i in range(number_of_subfolders):
        for j in range(segment_size):
            original_vido_id = subset.loc[i * segment_size + j]["original_vido_id"]
            new_original_vido_id = f"{original_vido_id}.{i + 1}"
            new_video_id = video_id
            new_frame_id = j + 1
            old_path = subset.loc[i * segment_size + j]["path"]
            new_path = f"{new_original_vido_id}/{new_frame_id}.jpg"
            new_labels = subset.loc[i * segment_size + j]["labels"]
            new_df.append([new_original_vido_id, new_video_id, new_frame_id, new_path, new_labels])
            
            if not os.path.exists(f"{new_dataset_path}/dataset/image/{new_original_vido_id}"):
                os.makedirs(f"{new_dataset_path}/dataset/image/{new_original_vido_id}")

            shutil.copy(f"{dataset_path}/dataset/image/{old_path}", f"{new_dataset_path}/dataset/image/{new_path}")
        
        video_id += 1
    
    for i in range(len(subset) % segment_size):
        original_vido_id = subset.loc[number_of_subfolders * segment_size + i]["original_vido_id"]
        new_original_vido_id = f"{original_vido_id}.{number_of_subfolders + 1}"
        new_video_id = video_id
        new_frame_id = i + 1
        old_path = subset.loc[number_of_subfolders * segment_size + i]["path"]
        new_path = f"{new_original_vido_id}/{new_frame_id}.jpg"
        new_labels = subset.loc[number_of_subfolders * segment_size + i]["labels"]
        new_df.append([new_original_vido_id, new_video_id, new_frame_id, new_path, new_labels])
    
        if not os.path.exists(f"{new_dataset_path}/dataset/image/{new_original_vido_id}"):
            os.makedirs(f"{new_dataset_path}/dataset/image/{new_original_vido_id}")

        shutil.copy(f"{dataset_path}/dataset/image/{old_path}", f"{new_dataset_path}/dataset/image/{new_path}")
    
    video_id += 1
        

new_train_df = pd.DataFrame(new_df, columns=["original_vido_id", "video_id", "frame_id", "path", "labels"])
new_train_df.to_csv(f"{new_dataset_path}/annotation/train.csv", sep=" ", index=False)
```
